```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from src.gimp.draw_rectangle import draw_rectangle
```

```python
green = (0.32, 0.74, 0.10)
purple = (0.5529412, 0.4627451, 0.8509804)
```

```python
def draw(y_):
    path_to_image_results = '../result/gimp_images/nn/square_parameters'
    name = 'jupyter_test.png'
    draw_rectangle(name=name, size=400, x=y_[4], y=y_[5], w=y_[6], h=y_[7], r=y_[0], g=y_[1], b=y_[2], a=y_[3], 
                   rotation=y_[8])
    return plt.imread('%s/%s' % (path_to_image_results, name))[:,:,:3]
```

```python
image = plt.imread('../data/image/hatch.png')[:,:,:3]
plt.imshow(image)
```

```python
drawned = draw([purple[0], purple[1], purple[2], 1., 0., 0.5, 1., 0.2, 0.5])
plt.imshow(drawned)
```

```python
diff = abs(image - drawned)
plt.imshow(diff)
```
